11. last_review - Дата последнего отзыва
12. reviews_per_month - Количество отзывов за месяц
13. calculated_host_listings_count - количество объявлений на один хостинг
14. availability_365 - количество дней, в течение которых список доступен для бронирования

#### У нас есть значение None в name и host_name
#### Посчитал что это это не прям критично заменю их на 'unspecified'

```python
# 

df.name.fillna('unspecified', inplace=True)
df.host_name.fillna('unspecified', inplace=True)

df.isnull().sum()
```

```python
df[df["last_review"].isnull() == True][:5]
```

#### Так как у нас 10000 нулл значении в review per month то здесь скорее всего просто не было вообще отзывов либо могла произойти ошибка при скрапинге.
#### Я посчитал нужным заменить их на ноль так как непонятна причина пропуска 
#### Так же и со столбцом last_review  заменю просто самой старой датой из датасета

```python
df['reviews_per_month'].fillna(0, inplace=True)
df['reviews_per_month'].isnull().sum()

df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

df['last_review'].fillna(df['last_review'].max(), inplace=True)
print(df.isnull().sum())
```

### Работа с выброосами в столбце прайс

```python
shared_room = df[df['room_type'] == 'Shared room']

shared_room[shared_room['price'] >= 1000]
```

```python
private_room = df[df['room_type'] == 'Private room']

private_room[private_room['price'] >= 3000]
```

```python
df[df['price'] == 0]
```

### С типом жилья private room посчитал что после 3000 долларов просто слишком завышено и скорее нереально
### С типом жилья shared room посчитал что после 1000 долларов тоже нереально
### И Удалим жилье где прайс стоит 0


```python
# Удаляем Shared room с ценой >= 1000
df.drop(df[(df['room_type'] == 'Shared room') & (df['price'] >= 1000)].index, inplace=True)

# Удаляем Private room с ценой >= 3000
df.drop(df[(df['room_type'] == 'Private room') & (df['price'] >= 3000)].index, inplace=True)

# Удаляем строки, где цена == 0
df.drop(df[df['price'] == 0].index, inplace=True)

df.reset_index(drop=True, inplace=True)

df.isnull().sum()
```
